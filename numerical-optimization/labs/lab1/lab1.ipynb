```python
plt.hist(np.random.uniform(low=-1, high=1, size=10**4), bins=100, density=1)
plt.axhline(y=0.5, color='r', linestyle='-')
plt.show()

plt.hist(np.random.uniform(low=-1, high=1, size=10**5), bins=100, density=1)
plt.axhline(y=0.5, color='r', linestyle='-')
plt.show()
```

```python
from scipy.stats import multivariate_normal
mu, cov = [5, 3], [[2, 0], [0, 1]]

x = np.linspace(0, 10, 100)
y = np.linspace(0, 6, 100)
X, Y = np.meshgrid(x, y)
pos = np.dstack((X, Y))

rv = multivariate_normal(mu, cov)
Z = rv.pdf(pos)

plt.contour(X, Y, Z, cmap='Reds', linewidths=3, antialiased=True)
plt.colorbar(label='density')
plt.scatter(*np.random.multivariate_normal(mu, cov, 10**4).T, alpha=0.5, s=0.5)
plt.show()
```

# Excercise 4

```python
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data[:, :2]
Y = iris.target

def pltscatter(color="red", **kwargs):
	plt.scatter(*X.T, color=color, edgecolors="black", **kwargs)
	plt.xlabel('sepal length (cm)')
	plt.ylabel('sepal width (cm)')
	plt.title('Iris')
```

```python
pltscatter()
plt.show()
```

```python
pltscatter()
plt.xlim(3, 9)
plt.ylim(1, 5)
plt.show()
```

```python
pltscatter()
plt.xticks(np.arange(3, 10))
plt.yticks(np.arange(1, 6))
plt.show()
```
