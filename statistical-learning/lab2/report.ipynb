```r
# Load necessary library
library(stats)

# Define degrees of freedom parameters
df1 <- 4
df2 <- 1000

# Generate x values for plotting
x <- seq(0, 5, length.out = 1000)

# Calculate the density of the F distribution
y <- df(x, df1, df2)

# Plot the F distribution
plot(x, y, type = "l", lwd = 2, col = "blue", 
     main = "F Distribution", xlab = "x", ylab = "Density")

# Add a legend
legend("topright", legend = paste("df1 =", df1, ", df2 =", df2), 
       col = "blue", lwd = 2, cex = 0.8)
```
