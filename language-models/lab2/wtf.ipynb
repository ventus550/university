```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
from torch.nn import functional as F
import random


model_name = 'eryk-mazus/polka-1.1b'
device = 'cuda'

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name).to(device)
```

```python
 vocab = tokenizer.get_vocab()
```

```python
tokenizer.decode([20775, 20775])
```

```python
for key, val in vocab.items():
	assert val == tokenizer.convert_tokens_to_ids(key)
```

```python
for key, val in vocab.items():
	assert key == tokenizer.convert_ids_to_tokens(val)
```

```python
for key, val in vocab.items():
	assert key == tokenizer.decode(val), (key, val, tokenizer.decode(val))
```

```python
input_ids = tokenizer("Ala ma kota", return_tensors='pt')['input_ids'].to(device)
# Generate logits for the next token
with torch.no_grad():
    outputs = model(input_ids)
    next_token_logits = outputs.logits[:, -1, :]  # Get logits for the last token

# Convert logits to probabilities
probs = F.softmax(next_token_logits, dim=-1).squeeze()

# Get tokens that start with "p" or "P" and set their probability to 0
for token_id in range(len(probs)):
    token = tokenizer.convert_ids_to_tokens(token_id)

    if not token or not token.startswith("p"): # or not token.startswith("P"):
        probs[token_id] = 0.0

# Renormalize probabilities
probs /= probs.sum()

torch.where(probs > 0)

# Sample from the filtered distribution
# filtered_token_id = torch.multinomial(probs, num_samples=1).item()
# filtered_token = tokenizer.decode([filtered_token_id])
# torch.where(probs > 0)[0]
A = "".join(tokenizer.batch_decode(torch.where(probs > 0)[0], skip_special_tokens=True))
B = "".join(tokenizer.convert_ids_to_tokens(torch.where(probs > 0)[0]))
C = "".join(tokenizer.decode(token_ids=torch.where(probs > 0)[0]))
```
